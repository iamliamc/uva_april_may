```python
# Run the following code to import the dataset to the df (dataframe)
# Do not change this cell!
%matplotlib inline
import pandas as pd
import requests
import io

csv = requests.get('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data').text
csv = csv.replace(' ', '')
df = pd.read_csv(io.StringIO(csv), header=None)
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 
              'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 
              'native-country', 'result']
df
```

#### Question 1 (1.0 points)
We will not need the *fnlwgt, education-num, capital-gain, capital-loss*, and *result* column. Remove them and then print the number of columns and rows, as well as the first few rows of the dataset.

```python
df = df.drop(['fnlwgt', 'education-num', 'capital-gain', 'capital-loss'], axis=1)
print("rows: %s, columns: %s" % df.shape)
df.head(2)
```

#### Question 2 (1.0 points)

#### Question 3 (1.0 points)
Find the mean hours per week worked (separately for each category) of those with marital status *Married-civ-spouse, Divorced, Separated* and *Never-married*. Print the results. Which group works more?

```python
def average_hours_worked(status):
    status_selection = df.loc[df['marital-status'] == status]
    print "%s worked an average of %s hours a week" % (status, (status_selection['hours-per-week'].mean()))

average_hours_worked("Married-civ-spouse")
average_hours_worked("Divorced")
average_hours_worked("Separated")
average_hours_worked("Never-married")
print "Married-civ-spouse works most"
```

#### Question 4 (1.0 points)
What is the mean age of those who work more than or equal to 40 hours a week, and those that work less than or equal to 20? Find and print the results.

```python
forty_plus_workers = df.loc[df['hours-per-week'] >= 40]
twenty_plus_workers = df.loc[df['hours-per-week'] >= 20]

print forty_plus_workers['age'].mean()
print twenty_plus_workers['age'].mean()
```

#### Question 5 (1.0 points)
Replace the Male/Female in the sex column with m/f, as appropriate.

```python
def change_sex_to_char(current):
    if current == "Male":
        return "m"
    elif current == "Female":
        return "f"
    else:
        raise TypeError
        

df['sex'] = df['sex'].apply(lambda x:change_sex_to_char(x))
df
```

#### Question 6 (2.0 points)
Find the mean worked hours per week by country, and then plot the top 10 most hardworking countries.

```python
df.groupby(by="native-country")['hours-per-week'].mean().sort_values(0, ascending=False).head(10).plot("bar")
```

#### Question 7 (2.0 points)
For everyone who is working in the United States ('United-States' in the dataset), find the mean age, the most common education level, and the most common marital status.

#### Question 8 (2.0 points)
Find the max, mean, and minimum capital gain in the United States, and find the number of people in the United States that have capital gains in the margins: [mean - 500, mean + 500]. So if the mean is 1000, find the people with capital gain from 500 to 1500.

```python
# we dropped capital gains in an earlier question create second intact dataframe
united_states_with_capital_gains = df_with_capital_gains[df_with_capital_gains['native-country'] == 'United-States']
max_cap = united_states_with_capital_gains['capital-gain'].max()
min_cap = united_states_with_capital_gains['capital-gain'].min()
mean_cap = united_states_with_capital_gains['capital-gain'].mean()
print "Max Capital Gains: %s, Min Capital Gains: %s, Mean Capital Gains: %s" % (max_cap, min_cap, mean_cap)

#get all capital gains less than mean + 500
cap_gains_less_than_max_margin = df_with_capital_gains.loc[df_with_capital_gains['capital-gain'] <= (mean_cap + 500)]
#get capital gains from those less than mean + 500 that are also greater than mean - 500
cap_gains_in_margin = cap_gains_less_than_max_margin.loc[cap_gains_less_than_max_margin['capital-gain'] >= (mean_cap - 500)]
cap_gains_in_margin
```
